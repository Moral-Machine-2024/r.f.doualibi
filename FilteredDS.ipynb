```python
import pandas as pd
import time

pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', None)

start_time = time.time()

df = pd.read_csv('SharedResponses.csv', chunksize = 100000)

for chunk in df: 
    filtered_chunk = chunk[chunk['UserCountry3'] == 'USA']
    display(filtered_chunk)
    break
end_time = time.time()
elapsed_time = end_time - start_time
print(f"The code took {elapsed_time:.2f} seconds to run.")
print(df)
```

```python
COUNTS OF ALL COUNTRIES IN THE USERCOUNTRY3 COLUMN (D0ES NOT WORK YET)
```

```python
import pandas as pd
import time

pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', None)

start_time = time.time()

# Specify data types for columns if needed
dtype = {'UserCountry3': str}  # Assuming the column UserCountry3 contains string values

df = pd.read_csv('SharedResponses.csv', chunksize=10, dtype=dtype)

# Initialize an empty dictionary to store counts
country_counts = {}

for chunk in df:
    # Count the occurrences of each unique country in the current chunk
    chunk_counts = chunk['UserCountry3'].value_counts()
    
    # Update the overall country counts dictionary with counts from the current chunk
    for country, count in chunk_counts.items():
        if country in country_counts:
            country_counts[country] += count
        else:
            country_counts[country] = count

# Convert the country counts dictionary to a DataFrame for easier display
country_counts_df = pd.DataFrame.from_dict(country_counts, orient='index', columns=['Count'])

# Display the counts of every country mentioned in the 'UserCountry3' column
print(country_counts_df)

end_time = time.time()
elapsed_time = end_time - start_time
print(f"The code took {elapsed_time:.2f} seconds to run.")
```
